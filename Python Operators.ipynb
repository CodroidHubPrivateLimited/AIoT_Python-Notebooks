# Python Operators

Operators are used to perform operations on variables and values.

Python divides the operators in the following groups:

- Arithmetic operators
- Assignment operators
- Comparison operators
- Logical operators
- Identity operators
- Membership operators
- Bitwise operators

## Arithmetic Operators

Arithmetic operators are used with numeric values to perform common mathematical operations:

### Addition

```python
var_1 = 10
var_2 = 20
print(var_1 + var_2)
```

### Subtraction

```python
var_1 = 10
var_2 = 20
print(var_1 - var_2)
```

### Multiplication

```python
var_1 = 10
var_2 = 20
print(var_1 * var_2)
```

### Division

```python
var_1 = 10
var_2 = 20
print(var_1 / var_2)
```

### Modulus

```python
var_1 = 10
var_2 = 20
print(var_1 % var_2)
```

### Exponentiation

```python
var_1 = 10
var_2 = 20
print(var_1 ** var_2)
```

### Floor Division

```python
var_1 = 10
var_2 = 20
print(var_1 // var_2)
```

## Identity Operators

```python
int_1 = 10
int_2 = 20
int_1 is int_2
```

```python
int_1 = 10
int_2 = 20
int_1 is not int_2
```

## Membership Operator

```python
str_1 = "abcd"
str_2 = "b"
str_2 in str_1
```

```python
str_1 = "abcd"
str_2 = "b"
str_2 not in str_1
```
