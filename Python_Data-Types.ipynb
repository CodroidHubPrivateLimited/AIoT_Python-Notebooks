## Text Type: String

A string is: 
- Sequence of characters
- Either as a literal constant or as some kind of variable
- Surrounded by either single quotation marks, or double quotation marks.

```python
# assigning 
name_1 = "My name is Tarun"
name_2 = 'My name is Somya'
# here the variable name_1 and ame_2 is different
print(name_1)
print(name_2)
```

- Assigning Multiline string to variables by just putting triple quotes

```python
introduction_1 = """Hello everyone,
My name is Karan,
I love python"""
print(introduction_1)
```

- Taking string as an input from user

```python
string_1 = "Hello World" # random format 
print(string_1.lower())  # converting to upper case
print(string_1.upper())  # converting to lower case
```

### String Formatting

As we all know that we can't concaticate number or intiger with string as it gives error to us. But we can do it using the format() function in python. It is a form of concatination in which we can concatenate numbers with strings.

format(): format function to concatinate string with a number

```python
time = 10
txt = "I came late to the class at {} AM" # here we take '{}' in a string where we have to add number
print(txt.format(time)) # passing variable contaning the number through function format()
```

```python
age = 10
str1 = "My age is: {}"
print(str1.format(age))
```

- Real Live representation of format function

```python
quantity = 3
item_no = 567
price = 49.95
my_order = "I want {} pieces of item number {} for {} dollars."
print(my_order.format(quantity, item_no, price))   
#   we have to put the parameteres in the sequence in which we have to allign them into that string
```

- Indexing on the values for allignment


```python
quantity = 3
item_no = 567
price = 49.95
my_order = "I want to pay {2} dollars for {0} pieces of item {1}."
print(my_order.format(quantity, item_no, price))
```

## Python Numeric Types

There are 3 types of numbers in python:
- Int
- Float
- Complex

```python
# Here we are assigning numbers to some variables of different numbers
x = 2     # int
y = 2.0   # float
z = 1j    # complex ( it is in the form of "ij" where "i" is an int and j is an imaginary number)

# now if we print the type of variables they are.
print(type(x))
print(type(y))
print(type(z))
```

### Type casting of Numbers

```python
# We can change the type of numbers just by putting its new type followed by the number
num = 10
num1 = str(num)
num2 = complex(num)  # it will changr the int and add "ij" to it to make it complex type
print(type(num2))
```

```python
# We can also create a number instantly and give it a type
number = int(5)
number1 = int(5.0)
number2 = complex(5j)
```
